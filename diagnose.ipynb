```python
# diagnose.py
import pandas as pd
import numpy as np

df = pd.read_csv("customer_data.csv")
print("Buy ratio in full dataset:", df["will_buy"].mean())
print()

X_train = np.load("X_train.npy")
y_train = np.load("y_train.npy")
X_test  = np.load("X_test.npy")
y_test  = np.load("y_test.npy")

print("y_train buy ratio:", y_train.mean())
print("y_test buy ratio :", y_test.mean())
print()
print("X_train min/max:", X_train.min(), X_train.max())
print("X_train sample (first sequence, first day):")
print(X_train[0][0])
print()
print("Any NaN in X_train?", np.isnan(X_train).any())
print("Any NaN in y_train?", np.isnan(y_train).any())
```

```python
# check_training.py
from tensorflow.keras.models import load_model
import numpy as np

model = load_model("lstm_model.h5")
X_test = np.load("X_test.npy")

preds = model.predict(X_test[:20], verbose=0).flatten()
print("First 20 raw predictions:")
print(preds.round(4))
print()
print("Min prediction:", preds.min())
print("Max prediction:", preds.max())
print("Std deviation:", preds.std())
```
